### In-class Exercise: Uniform Distribution - Sample Mean and Standard Deviation

This code calculates the mean, z, and standard deviation, sigma, of a set of generated N_max random numbers uniformly distributed in a domain $[a,b]$ where $\textit {a}$ and $\textit {b}$ could be real numbers or integers, using the package $ \bf{random} $ (https://docs.python.org/3/library/random.html). Mean and standard deviation are calculated using the $ \bf {numpy} $ package (https://numpy.org). 

<ul> 
    <li> The minimum sample size is 3.</li>
    <li> For this exercise you will vary N_max from 10 to about 10,000 (in steps of 200, may be).</li>
</ul>

##### Notes.

    <li> For this exercise you will vary N_max from 10 to about 10,000 (in steps of 200, may be).</li>

```python
import random
import numpy as np
import matplotlib.pyplot as plt

mean = np.array([])
stdev = np.array([])
sample_size = np.array([])

# calculate mean and standard deviatiion for uniform random numbers in the range (0,1) of various sizes ranging 
# from 3 to N_max.
#calculations below use N_max = 1000. Real random variable belongs to the domain [a,b]=[0,1]

a = 0
b = 1

mean_t = (b+a)/2
stdev_t = np.sqrt((b-a)**2/12)

#varies the sample size from 3 to N_max. 

for i in range (3, 1000):
    y=[random.random() for _ in range(i)]
    
    z = np.average(y)
    sigma = np.std(y)
    mean=np.append(mean, z)
    stdev=np.append(stdev, sigma)
    sample_size=np.append(sample_size,i)
print('theoretical mean=', mean_t, 'and theoretical standard deviation=', stdev_t)

#calculate absolute value of the difference between sample mean and theoretical mean

diff_mean = np.abs(mean-mean_t)

plt.figure(figsize=(25,16))
plt.plot(sample_size, mean, label = 'sample mean', color='blue')
plt.plot(sample_size, stdev, label = 'standard deviation', color='magenta')
plt.plot(sample_size, diff_mean, label = 'difference', color='green')
plt.legend(fontsize = 20)

```

```python
import random
import numpy as np
import matplotlib.pyplot as plt

mean = np.array([])
stdev = np.array([])
sample_size = np.array([])

#calculate mean and standard deviatiion for uniform integer random numbers in the range (0,100) of various sizes ranging 
# from 3 to N_max.
#calculations below use N_max = 1000. Integer random variable belongs to the domain [a,b]=[0,100]

a = 0
b = 100

mean_t = (b+a)/2
stdev_t = np.sqrt((b-a)**2/12)

for i in range (3, 1000):
    
    # if you would like to create a random integer between 0 to 100, use this command instead
    y = [random.randrange(0,101) for _ in range(i)]
    z = np.average(y)
    sigma = np.std(y)
    mean = np.append(mean, z)
    stdev = np.append(stdev, sigma)
    sample_size = np.append(sample_size,i)
    
print('theoretical mean=', mean_t, 'and theoretical standard deviation=', stdev_t)

#calculate absolute value of the difference between sample mean and theoretical mean
diff_mean = np.abs(mean-mean_t)

plt.figure(figsize=(25,16))
plt.plot(sample_size, mean, label = 'sample mean', color='black')
plt.plot(sample_size, stdev, label = 'standard deviation', color='red')
plt.plot(sample_size, diff_mean, label = 'difference', color='green')
plt.legend(fontsize = 20)
```
